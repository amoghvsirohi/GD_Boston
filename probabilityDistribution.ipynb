**Loading relevant libraries**

```
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
from scipy.stats import bernoulli, norm, laplace, expon, poisson
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

**2. i and ii**

```
Zbern = 0
Znorm = 0
Zlaplace = 0
Zexpon = 0
Zpoisson = 0

for i in range(1001):
  Zbern = Zbern + bernoulli.pmf(i/1000, 0.7)
  Znorm = Znorm + norm.pdf(i/1000)
  Zlaplace = Zlaplace + laplace.pdf(i/1000)
  Zexpon = Zexpon + expon.pdf(i/1000)
  Zpoisson = Zpoisson + poisson.pmf(i/1000,0.3)

bern_list = []
norm_list = []
laplace_list = []
expon_list = []
poisson_list = []

bern = bernoulli(0.7)
poi = poisson(0.3)
for i in range(1001):
  bern_list.append(bern.pmf(i/1000)/Zbern)
  norm_list.append(norm.pdf(i/1000)/Znorm)
  laplace_list.append(laplace.pdf(i/1000)/Zlaplace)
  expon_list.append(expon.pdf(i/1000)/Zexpon)
  poisson_list.append(poi.pmf(i/1000)/Zpoisson)

print('Z Bernoulli:',Zbern)  
print('U Bernoulli:',bern_list)

print('Z Norm:',Znorm)
print('U Norm:',norm_list)

print('Z Laplace:',Zlaplace)
print('U Laplace:',laplace_list)

print('Z Expon:',Zexpon)
print('U Expon:',expon_list)

print('Z Poisson:',Zpoisson)
print('U Poisson:',poisson_list) 
```

**2. a.**

```
storage = []

for i in range(1001):
  storage.append(i/1000)

plt.plot(storage,bern_list)
plt.xlabel('Distribution')  
plt.ylabel('Bernoulli') 
plt.show()

plt.plot(storage,norm_list)
plt.xlabel('Distribution')  
plt.ylabel('Norm') 
plt.show()

plt.plot(storage,laplace_list)
plt.xlabel('Distribution')  
plt.ylabel('Laplace') 
plt.show()

plt.plot(storage,expon_list)
plt.xlabel('Distribution')  
plt.ylabel('Expon') 
plt.show()

plt.plot(storage,poisson_list)
plt.xlabel('Distribution')  
plt.ylabel('Poisson') 
plt.show()
```

```
def kl_divergence(u,x):
	dkl=0
	for i in range(len(u)):
		if x[i] != 0 and u[i] !=0:
			dkl += u[i] * np.log(u[i]/x[i])
	return dkl  

print("**********KL(U||X)**************")
print('KL(U||X) Bernoulli:',kl_divergence(storage,bern_list))
print('KL(U||X) Norm:',kl_divergence(storage,norm_list))
print('KL(U||X) Laplace:',kl_divergence(storage,laplace_list))
print('KL(U||X) Expon:',kl_divergence(storage,expon_list))
print('KL(U||X) Poisson:',kl_divergence(storage,poisson_list))
print("**********KL(X||U)**************")
print('KL(X||U) Bernoulli',kl_divergence(bern_list,storage))
print('KL(X||U) Norm',kl_divergence(norm_list,storage))
print('KL(X||U) Laplace',kl_divergence(laplace_list,storage))
print('KL(X||U) Expon',kl_divergence(expon_list,storage))
print('KL(X||U) Poisson',kl_divergence(poisson_list,storage))
```
